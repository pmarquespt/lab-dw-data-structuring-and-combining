```python
import pandas as pd

data = pd.read_csv('cleaned_data.csv')
data.head()
```

```python
data2 = pd.read_csv('https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file2.csv')
data2 = pd.DataFrame(data)
data2.head()
```

```python
# 1 - Column names should be in lower case
data2.columns = data.columns.str.lower()
data2
```

```python
# 2 - *White spaces in column names should be replaced by `_`
data2.columns = data.columns.str.replace(' ', '_')
data2
```

```python
# 3 -`st` could be replaced for `state`

data2.rename(columns={'st': 'state'}, inplace=True)
data2
```

```python
# 1 Gender column contains various inconsistent values such as "F", "M", "Femal", "Male", "female", which need to be standardized, for example, to "M" and "F".*

data2["gender"].unique() # first we nead to see how many different values there are
```

```python
data2['gender'].fillna('unknown', inplace=True)  # After getting some errors we came to the conclusion that it would be necessary to fill in the NaN fields


data2['gender'] = data2['gender'].str.lower()  # In order to standardize, we change all words to lowercase letters.


# And then we replace all words that start with 'f' to 'F'  all words that start with 'm' to 'M'

data2.loc[data2['gender'].str.startswith('f'), 'gender'] = 'F'
data2.loc[data2['gender'].str.startswith('m'), 'gender'] = 'M'
data2.loc[data2['gender'].str.startswith('n'), 'gender'] = 'M'


data2["gender"].unique()
```

```python
import pandas as pd

data = pd.read_csv("https://raw.githubusercontent.com/data-bootcamp-v4/data/main/marketing_customer_analysis_clean.csv")
df_1 = pd.DataFrame(data)   

df_1.head()
```

```python
# 1. You work at the marketing department and you want to know which sales channel brought the most sales in terms of total revenue. 
# Using pivot, create a summary table showing the total revenue for each sales channel (branch, call center, web, and mail).
```

```python
resume = df_1.pivot_table(values='total_claim_amount', index='sales_channel', aggfunc='sum')
resume 
```

```python
# just interesting to see

resume.plot(kind='bar', figsize=(10, 6))
```

```python
# Round the total revenue to 2 decimal points.  Analyze the resulting table to draw insights.

resume = df_1.pivot_table(values='total_claim_amount', index='sales_channel', aggfunc='sum')


resume['total_claim_amount'] = resume['total_claim_amount'].apply(lambda x: round(x, 2))

resume
```

```python
# 2. Create a pivot table that shows the average customer lifetime value per gender and education level. Analyze the resulting table to draw insights.

customer_value_pivot = df_1.pivot_table(values='customer_lifetime_value', index='gender', columns='education', aggfunc='mean')

customer_value_pivot
```

```python
# just interesting to see

customer_value_pivot.plot(kind='bar', figsize=(10, 6))
```
